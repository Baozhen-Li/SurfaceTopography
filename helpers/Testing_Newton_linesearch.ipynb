```python
import sys
import os
import numpy as np

import scipy.optimize
sys.path.append(os.path.join(os.getcwd(), "../PyPyContact/Tools/"))
```

```python
def fun(x):
    return .5*x[0]**2 + x[0]* np.cos(x[1])

def jac(x):
    return np.array([[x[0] + np.cos(x[1])],
                     [-x[0] * np.sin(x[1])]])

def hess(x):
    return np.array([[           1.,        -np.sin(x[1])],
                     [-np.sin(x[1]), -x[0] * np.cos(x[1])]])
```

```python
from NewtonLineSearch import newton_linesearch
```

```python
fname = "../PyPyContact/Tools/NewtonLineSearch.py"
with open(fname) as filehandle:
    content = ''.join((line for line in filehandle))
    exec(content)
```

```python
x0 = np.array([1.,1.])
result = scipy.optimize.minimize(fun, x0=x0, method=newton_linesearch,
                                 jac=jac, hess=hess, tol=1.e-16,
                                 options={'store_iterates': 'iterate'})
print(" k    f(x_k)   |∇f(x_k))| α_k    τ_k")
for (i, it) in enumerate(result.iterates):
    print("{0:>2} {1:+.4e} {2:.4e} {3:>2} {4:.4e}".format(i, it.fun, np.linalg.norm(it.jac), it.alpha, it.tau))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
iterates = np.array([iterate.x for iterate in result.iterates])
x, y = iterates[:, 0], iterates[:, 1]
xg, yg = np.linspace(-2, 2, 51), np.linspace(-6, 6, 51)
X, Y = np.meshgrid(xg,yg)
plt.contour(X, Y, fun([X, Y]))
plt.plot(x, y)
plt.figure()
xg, yg = np.linspace(-.3, 1.1, 51), np.linspace(-.95, 3.5, 51)
X, Y = np.meshgrid(xg,yg)
plt.contour(xg, yg, fun([X, Y]))
plt.plot(x, y)
```
